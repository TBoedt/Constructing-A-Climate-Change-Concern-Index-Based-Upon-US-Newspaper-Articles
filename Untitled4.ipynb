```python
v1 = v1[v1["was_I_retarded?"] != ""]
v2 = v2[v2["was_I_retarded?"] != ""]
v3 = v3[v3["was_I_retarded?"] != ""]
```

```python
v = pd.concat([v1, v2, v3]).reset_index(drop = True)
v = v.drop("level_0", axis = 1)
```

```python
v.loc[v["was_I_retarded?"] == "Y", "was_I_retarded?"] = "Yes"
v.loc[v["was_I_retarded?"] == "N", "was_I_retarded?"] = "No"
```

```python
table = table.merge(v[["Link", "was_I_retarded?"]], on="Link", how="left")
```

```python
table.loc[table["was_I_retarded?"].isna(), "was_I_retarded?"] = "Yes"
```

```python
table.loc[table["was_I_retarded?"] == "Yes", "Final_Sentiment_Label"] = table["Sentiment_Label"]
```

```python
r = table[(table["Final_Sentiment_Label"] == "TBD") & ((table["Final_Climate_Change_Level_Label"] == "High") | (table["Final_Climate_Change_Level_Label"] == "Medium"))].reset_index(drop = True)
```

```python
for h in range(0, len(r)):
    print(r.loc[h]["Link"])
    print(r.loc[h]["Text"])
    
    print("\n")
    print("T: "+r.loc[h]["Sentiment_Label"] )
    print("R: "+r.loc[h]["Sentiment_Label_R"] )
    r.loc[h, "Final_Sentiment_Label"] = input("Sentiment: ")
    print("\n")
```

```python
for s in table["Link"]:
    if (r["Link"] == s).any():
        print(int(r[r["Link"] == s]["Final_Sentiment_Label"]))
        table.loc[table["Link"] == s, "Final_Sentiment_Label"] = int(r[r["Link"] == s]["Final_Sentiment_Label"])
```

```python
table["Final_Sentiment_Label"] = table["Final_Sentiment_Label"].astype(int)
```
