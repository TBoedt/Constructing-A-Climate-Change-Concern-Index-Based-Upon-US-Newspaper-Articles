## 1.1. Events

### Extreme Weather US

```python
Extreme_Weather_Events_US = pd.read_csv("events-US-1980-2023.csv").reset_index()
Extreme_Weather_Events_US.columns = list(Extreme_Weather_Events_US.loc[0])
Extreme_Weather_Events_US = Extreme_Weather_Events_US.drop([0]).reset_index(drop = True)

#Transform the dates to actual date type
Extreme_Weather_Events_US["Begin Date"] = pd.to_datetime(Extreme_Weather_Events_US["Begin Date"], format="%Y%m%d")
Extreme_Weather_Events_US["End Date"] = pd.to_datetime(Extreme_Weather_Events_US["Begin Date"], format="%Y%m%d")

#only select events that started after or in 2016 and ended before or in 2020
Extreme_Weather_Events_US = Extreme_Weather_Events_US[(Extreme_Weather_Events_US["Begin Date"].dt.year >= 2016) & 
                          (Extreme_Weather_Events_US["End Date"].dt.year <= 2020)].reset_index(drop = True)

#transform string of numbers to numbers
Extreme_Weather_Events_US["Deaths"] = pd.to_numeric(Extreme_Weather_Events_US["Deaths"], downcast='integer')
Extreme_Weather_Events_US["Total CPI-Adjusted Cost (Millions of Dollars)"] = pd.to_numeric(Extreme_Weather_Events_US["Total CPI-Adjusted Cost (Millions of Dollars)"], downcast='integer')

Extreme_Weather_Events_US
```

```python
#Look at the events with the highest amount of deaths (Size)
Extreme_Weather_Events_US.sort_values(by = "Deaths", ascending = False).head(10)
```

```python
#Look at the events with the highest amount of Total CPI-Adjusted Cost (Size)
Extreme_Weather_Events_US.sort_values(by = "Total CPI-Adjusted Cost (Millions of Dollars)", ascending = False).head(10)
```
