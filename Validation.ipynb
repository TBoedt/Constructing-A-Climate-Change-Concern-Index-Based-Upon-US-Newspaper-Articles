### Extreme Weather US

```python
Extreme_Weather_Events_US = pd.read_csv("events-US-1980-2023.csv").reset_index()
Extreme_Weather_Events_US.columns = list(Extreme_Weather_Events_US.loc[0])
Extreme_Weather_Events_US = Extreme_Weather_Events_US.drop([0]).reset_index(drop = True)

#Transform the dates to actual date type
Extreme_Weather_Events_US["Begin Date"] = pd.to_datetime(Extreme_Weather_Events_US["Begin Date"], format="%Y%m%d")
Extreme_Weather_Events_US["End Date"] = pd.to_datetime(Extreme_Weather_Events_US["Begin Date"], format="%Y%m%d")

#only select events that started after or in 2016 and ended before or in 2020
Extreme_Weather_Events_US = Extreme_Weather_Events_US[(Extreme_Weather_Events_US["Begin Date"].dt.year >= 2016) & 
                          (Extreme_Weather_Events_US["End Date"].dt.year <= 2020)].reset_index(drop = True)

#transform string of numbers to numbers
Extreme_Weather_Events_US["Deaths"] = pd.to_numeric(Extreme_Weather_Events_US["Deaths"], downcast='integer')
Extreme_Weather_Events_US["Total CPI-Adjusted Cost (Millions of Dollars)"] = pd.to_numeric(Extreme_Weather_Events_US["Total CPI-Adjusted Cost (Millions of Dollars)"], downcast='integer')

Extreme_Weather_Events_US
```

```python
driver = webdriver.Chrome()
event = []
event_link = []
category = []
year = []
page_or_sub = []

for link in list(Events_df["Link Category"]):
    driver.get(link)
    try:
        subc = WebDriverWait(driver, 1).until(EC.presence_of_element_located((By.ID, "mw-subcategories")))
        subc = subc.find_elements(By.CLASS_NAME, "CategoryTreeItem")
        for sub in subc:
            sub_info = WebDriverWait(sub, 1).until(EC.presence_of_element_located((By.TAG_NAME, "a")))
            event.append(sub_info.text)
            event_link.append(sub_info.get_attribute('href'))
            category.extend(list(Events_df.loc[Events_df["Link Category"] == link, "Category"]))
            year.extend(list(Events_df.loc[Events_df["Link Category"] == link, "Year"]))
            page_or_sub.append("SubCategory")
            
    except:
        pass
    
    pages = WebDriverWait(driver, 1).until(EC.presence_of_element_located((By.CLASS_NAME, "mw-category")))
    pages = pages.find_elements(By.TAG_NAME, "a")
    for page in pages:
        event.append(page.text)
        event_link.append(page.get_attribute('href'))
        category.extend(list(Events_df.loc[Events_df["Link Category"] == link, "Category"]))
        year.extend(list(Events_df.loc[Events_df["Link Category"] == link, "Year"]))
        page_or_sub.append("Page")

driver.quit()
Events_df = pd.DataFrame({"Category" : category, "Event" : event, "Link Event" : event_link, 
                          "Year" : year, "Sub or Page" : page_or_sub})


#SubCategory_Research = Events_df[Events_df["Sub or Page"] == "SubCategory"].reset_index(drop = True)
```
